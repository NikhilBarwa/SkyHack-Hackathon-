```python
def convert_mixed_datetime(date_column):
    # Attempt to parse MM/DD/YYYY format first
    format1 = pd.to_datetime(date_column, format='%m/%d/%Y %H:%M', errors='coerce')
    
    # For rows where format1 failed (NaT), attempt to parse DD-MM-YYYY format
    format2 = pd.to_datetime(date_column, format='%d-%m-%Y %H:%M', errors='coerce')
    
    # Use format1 where it succeeded, otherwise use format2
    return format1.fillna(format2)
```

```python
for col in ['call_start_datetime', 'agent_assigned_datetime', 'call_end_datetime']:
    df_calls[col] = convert_mixed_datetime(df_calls[col])
```
